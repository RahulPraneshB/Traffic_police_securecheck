```python
print(data.isnull().sum())

a = data[(data['search_conducted'] == 0)]
display(a)

b = data[(data['search_conducted'] == 0 & data['search_type'].isnull())]
display(b)

c = b[b['search_type'].isnull()]
display(c)

b_so = c[c['stop_outcome'] == 'Arrest']
display(b_so)

print(b_so['violation'].unique())

a_filtered = a[a['search_type'].notna()]
display(a_filtered)

print(data['driver_gender'].unique())
print(data['country_name'].unique())
print(data['driver_race'].unique())
print(data['violation_raw'].unique())
print(data['violation'].unique())
print(data['stop_outcome'].unique())
print(data['search_type'].unique())
```

```python
condition = (data['search_conducted'] == False)
data.loc[condition, 'search_conducted'] = True
display(data)
```

```python
data[(data['search_type'].isnull())]
```

```python
data.loc[(data['search_type'].isnull()) & (data['violation_raw'].isin(['Drunk Driving', 'Other'])), 'search_type'] = 'Frisk'
data.loc[data['search_type'].isnull(), 'search_type'] = 'Vehicle Search'
display(data)
```

```python
print(data.isnull().sum())
```

```python
data['stop_date'] = pd.to_datetime(data['stop_date'])
data['stop_time'] = pd.to_datetime(data['stop_time'])
display(data)
```

```python
dtype = {
    'stop_date': types.DATETIME(),
    'stop_time': types.DATETIME(),
    'country_name': types.VARCHAR(255),
    'driver_gender': types.VARCHAR(10),
    'driver_age_raw': types.INTEGER(),
    'driver_age': types.INTEGER(),
    'driver_race': types.VARCHAR(32),
    'violation_raw': types.VARCHAR(255),
    'violation': types.VARCHAR(255),
    'search_conducted': types.BOOLEAN(),
    'search_type': types.VARCHAR(255),
    'stop_outcome': types.VARCHAR(255),
    'is_arrested': types.BOOLEAN(),
    'stop_duration': types.VARCHAR(64),
    'drugs_related_stop': types.BOOLEAN(),
    'vehicle_number': types.VARCHAR(64)
}
```

```python
data.to_sql(
    name='trafficlogs',
    con=engine,
    if_exists='replace',  # or 'append' if the table already exists
    index=False,
    dtype=dtype)
```
